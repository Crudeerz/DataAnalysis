## Split dataframes to separate frames
Get the values from each axis depending on PrimaryKey


```python
def parse(file, PK):
    
    a = file[file.PrimaryKey == PK]
    
    return a

Xaxis = parse(XYZ21_LOAD, 1)  
Yaxis = parse(XYZ21_LOAD, 2)
Zaxis = parse(XYZ21_LOAD, 3)

Xaxis1 = parse(XYZ21_LOAD1, 1)
Yaxis1 = parse(XYZ21_LOAD1, 2)
Zaxis1 = parse(XYZ21_LOAD1, 3)


```

* Return the time of sampling, the length of the file devided by 2


```python
#Take the file that has been resolved earlier and show the length and time of the reading
#tags = how many tags are read in the same csv file
#sTime = the sampling setting time
def fileSize(file, tags, sTime):
    l = len(file)
    approxLengthofTagFrame = l // tags
    approxTimeofReading = (approxLengthofTagFrame // 2) //60
    print('around', approxLengthofTagFrame, 'readings per tag, and', approxTimeofReading, 'minutes of reading')



fileSize(XYZ21_LOAD1, 3, 0.5)
```

## Draw each axis graph
* Prepare graphical axis for plot

```python
def draw(axisFrame):
    #plt.style.use('_mpl-gallery')

    GraphX = np.arange(len(axisFrame))
    plt.figure(figsize=(10, 5))
    plt.plot(GraphX, axisFrame.Value)
    plt.xlabel('Readings')
    plt.ylabel('LOAD in %')
    plt.show()
    print('insamlingstid = ca', ((len(axisFrame)) // 2) // 60,'min')
    #print(axisFrame)


#print('X-Axis')
draw(Xaxis1)
#print('Y-Axis')
draw(Yaxis1)
#print('Z-Axis')
draw(Zaxis1)
```

## Draw each axis on same graph

```python

def drawAll(axisFrames):
    GraphX = np.arange(len(axisFrames[0]))
    GraphY = np.arange(len(axisFrames[1]))
    GraphZ = np.arange(len(axisFrames[2]))
    plt.figure(figsize=(10, 10))
    plt.plot(GraphX, axisFrames[0].Value)
    plt.plot(GraphY, axisFrames[1].Value)
    plt.plot(GraphZ, axisFrames[2].Value)
    plt.legend(['XM21_LOAD', 'YM21_LOAD', 'ZM21_LOAD'])
    plt.xlabel('Readings')
    plt.ylabel('LOAD in %')
    plt.show()

drawAll([Xaxis1, Yaxis1, Zaxis1])
```
